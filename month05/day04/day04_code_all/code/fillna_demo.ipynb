```python
import pandas as pd
import numpy as np
```

```python
dates=pd.date_range('20130101',periods=6)
df=pd.DataFrame(data=np.arange(24).reshape((6,4)),
               index=dates,columns=list('ABCD'))
df.iloc[0,1]=np.NAN
df.iloc[1,2]=np.NAN
df
```

```python
# 如何检测缺失值
df.isnull()
df.notnull()
```

```python
# 每一行的缺失值的数量
# df1=df.isnull().sum(axis=1)
# df1
# 每一列的缺失值的数量
df2=df.isnull().sum(axis=0)
print(df2)
# 整组数据缺失值的情况
print(df2.sum())
```

```python
# 处理缺失数据的第一种方法：删除法
# 缺失值所在的行
df3=df.dropna(axis=0,how='any')
df3
```

```python
# 删除缺失值所在的列
df4=df.dropna(axis=1,how='any')
df4
```
