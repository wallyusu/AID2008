```python
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = 'SimHei' ## 设置中文显示
plt.rcParams['axes.unicode_minus'] = False ## 正常显示符号
import numpy as np
# 创建一个画布  8*100 长 6*100宽
fig=plt.figure(figsize=(8,6),dpi=100,facecolor='y')
# 查看默认的像素大小
# fig.dpi
# 创建x轴数据 等差数组（起始值，结束值，num）
data=np.linspace(0,4*np.pi,100)
y=np.sin(data)
# 绘制函数图形
plt.plot(data,y,color='c',linestyle='--')
# 添加x轴，y轴标签
plt.xlabel('data')
plt.ylabel('sin函数')
# 添加标题
plt.title('sin曲线')
plt.savefig(fname='img/sin曲线.png')
# 显示图形
plt.show()
```
