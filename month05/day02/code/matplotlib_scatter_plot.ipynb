```python
import numpy as np
import matplotlib.pyplot as plt
# 设置打印内容全部输出
np.set_printoptions(threshold=np.inf)  # 设置打印的内容足够大
plt.rcParams['font.sans-serif'] = 'SimHei' ## 设置中文显示
plt.rcParams['axes.unicode_minus'] = False ## 正常显示符号
# 读取国民生产总值季度数据
data = np.load('data/国民经济核算季度数据.npz',allow_pickle=True)
# data.files
# 提取数据
values = data['values']
# 提取列名称
columns = data['columns']
```

```python
# 绘制2000年-2017年各产业国民经济生产总值散点图
# 创建一个画布
plt.figure(figsize=(9,6),dpi=100)
# 绘制第一产业散点图
plt.scatter(values[:,0],values[:,3],marker='*',c='r')
# 绘制第二产业散点图
plt.scatter(values[:,0],values[:,4],marker='s',c='g')
# 绘制第三产业散点图
plt.scatter(values[:,0],values[:,5],marker='.',c='b')
# 添加标签
plt.xlabel('序号',fontsize=15)
plt.ylabel('生产总值',fontsize=15)
# 添加标
plt.title('2000年-2017年各产业国民经济生产总值散点图',fontsize=20)
# 添加图例
plt.legend(['第一产业','第二产业','第三产业'])
# 存储图形
plt.savefig('img/散点图.jpg')
# 显示图形
plt.show()
```

```python
# 绘制2000-2017年各产业以及各行业国民生产总值折线图
label1 = ['第一产业','第二产业','第三产业'] 
label2 = ['农业','工业','建筑','批发','交通','餐饮','金融','房地产','其他']
# 创建画布
fig2 = plt.figure(figsize=(15,10),dpi=100)
# 创建子图1 创建并选中
fig2.add_subplot(2,1,1)
# 绘制三大产业折线图
plt.plot(values[:,0],values[:,3],'b-',
        values[:,0],values[:,4],'r-',
        values[:,0],values[:,5],'g--')
plt.ylabel('生产总值(亿元)')
plt.legend(label1)
plt.title('2000-2017年各产业以及各产业国民生产总值折线图')
# 创建子图2
fig2.add_subplot(2,1,2)
# 绘制九大行业折线图
plt.plot(values[:,0],values[:,6],'b-',
        values[:,0],values[:,7],'r-',
        values[:,0],values[:,8],'g--',
        values[:,0],values[:,9],'k-',
        values[:,0],values[:,10],'c-',
        values[:,0],values[:,11],'y--',
        values[:,0],values[:,12],'m-',
        values[:,0],values[:,13],'r-',
        values[:,0],values[:,14],'c--'
        )
plt.xlabel('年份')
plt.ylabel('生产总值(亿元)')
plt.legend(label2)
plt.title('2000-2017年各产业以及各行业国民生产总值折线图')
# 设置x轴刻度，位置，取值
plt.xticks(range(0,70,4),values[range(0,70,4),1],rotation=45)
# 保存图形
plt.savefig('img/折线图2.jpg')
# 显示图形
plt.show()
```
