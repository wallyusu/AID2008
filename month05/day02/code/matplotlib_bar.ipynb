```python
import numpy as np
import matplotlib.pyplot as plt
# 设置打印内容全部输出
np.set_printoptions(threshold=np.inf)  # 设置打印的内容足够大
plt.rcParams['font.sans-serif'] = 'SimHei' ## 设置中文显示
plt.rcParams['axes.unicode_minus'] = False ## 正常显示符号
# 读取国民生产总值季度数据
data = np.load('data/国民经济核算季度数据.npz',allow_pickle=True)
# data.files
# 提取数据
values = data['values']
# 提取列名称
columns = data['columns']
```

```python
label1 = ['第一产业','第二产业','第三产业'] 
label2 = ['农业','工业','建筑','批发','交通','餐饮','金融','房地产','其他']
```

```python
a = []
```

```python
# 创建画布
fig = plt.figure(figsize=(12,12),dpi=100)
# 子图1
fig.add_subplot(2,2,1)
plt.bar(range(3),values[0,3:6],width=0.6,color=['r','g','b'],alpha=0.2) # alpgha 透明度
plt.xlabel('产业')
plt.ylabel('生产总值(亿元)')
plt.title('2000年第一季度国民生产总值构成柱状图')
plt.xticks(range(3),label1)
for a,b in zip(range(3),values[0,3:6]):
    plt.text(a,b,'%.1f'%b,ha='center',va='bottom',fontsize=10)
# 子图2
fig.add_subplot(2,2,2)
plt.bar(range(3),values[-1,3:6],width=0.6,color=['r','g','b'])
plt.xlabel('产业')
plt.ylabel('生产总值(亿元)')
plt.title('2017年第一季度国民生产总值构成柱状图')
plt.xticks(range(3),label1)
for a,b in zip(range(3),values[-1,3:6]):
    plt.text(a,b,'%.1f'%b,ha='center',va='bottom',fontsize=10)
# 子图3
fig.add_subplot(2,2,3)
plt.bar(range(9),values[0,6:],width=0.5,color=['b','r','g','k','m','c','y','m','b'])
plt.xlabel('行业')
plt.ylabel('生产总值')
plt.title('2000年第一季度国民生产总值行业构成柱状图')
plt.xticks(range(9),label2)
for a,b in zip(range(9),values[0,6:]):
    plt.text(a,b,'%.1f'%b,ha='center',va='bottom',fontsize=10)
# 子图4
fig.add_subplot(2,2,4)
plt.bar(range(9),values[-1,6:],width=0.5,color=['b','r','g','k','m','c','y','m','b'])
plt.xlabel('行业')
plt.ylabel('生产总值')
plt.title('2017年第一季度国民生产总值行业构成柱状图')
plt.xticks(range(9),label2)
for a,b in zip(range(9),values[-1,6:]):
    plt.text(a,b,'%.1f'%b,ha='center',va='bottom',fontsize=10)
#保存图形 
plt.savefig('img/2000和2017第一季度柱状图.jpg')
plt.show()
```

```python
explode1 = [0.01,0.03,0.01]
explode2 = [0.01,0.01,0.01,0.01,0.01,0.01,0.01,0.01,0.01]

fig2 = plt.figure(figsize=(12,12),dpi=100)
# 子图1 2000年第一季度国民生产总值产业分布饼图
fig2.add_subplot(2,2,1)
plt.pie(x=values[0,3:6],labels=label1,explode=explode1,autopct='%4.1f%%')
plt.title('2000年第一季度国民生产总值产业分布饼图')
# 子图2 2017年第一季度国民生产总值产业分布饼图
fig2.add_subplot(2,2,2)
plt.pie(x=values[-1,3:6],labels=label1,explode=explode1,autopct='%4.1f%%')
plt.title('2017年第一季度国民生产总值产业分布饼图')
# 子图3 2000年第一季度国民生产总值行业分布饼图
fig2.add_subplot(2,2,3)
plt.pie(x=values[0,6:],labels=label2,explode=explode2,autopct='%4.1f%%')
plt.title('2000年第一季度国民生产总值行业分布饼图')
# 子图4 2017年第一季度国民生产总值行业分布饼图
fig2.add_subplot(2,2,4)
plt.pie(x=values[-1,6:],labels=label2,explode=explode2,autopct='%4.1f%%')
plt.title('2017年第一季度国民生产总值行业分布饼图')
# 保存图形
plt.savefig('img/饼图.jpg')
plt.show()
```
