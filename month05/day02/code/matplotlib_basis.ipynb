```python
# 导入绘图库
import matplotlib.pyplot as plt
import numpy as np
# 创建x轴数据
data=np.arange(0,1.1,0.1)
data
# 绘制图形
plt.plot(data,data**2)
plt.plot(data,data**4)
# 添加绘图主体信息
# 添加x轴，y轴标签
plt.xlabel(s='x',fontsize=18)
plt.ylabel(s='y',fontsize=18)
# 添加标题
plt.title(s='lines',fontsize=25)
# 设置坐标范围和刻度
plt.xlim(0,1)
plt.ylim(0,1)
# plt.xticks([0,0.3,0.5,0.7,0.9],fontsize=16)
# plt.yticks([0,0.2,0.4,0.6,0.8],fontsize=16)
# 添加图例
plt.legend(labels=['y=x^2','y=x^4'],loc=2,fontsize=15)
# 保存图形
plt.savefig(fname='img/y=x^2&y=x^4.png')
# 显示图形
plt.show()
```
