```python
import pandas as pd
import numpy as np
# 从用户表中读取用户数据
users=pd.read_table('DS_day03/code/data/users.dat',
                   header=None,
                   names=['UserID','Gender','Age','Occupation','Zip-code'],
                   sep='::',engine='python')
users.head()
```

```python
type(users)
```

```python
# 从电影表中读取电影详情
movies=pd.read_table('DS_day03/code/data/movies.dat',
                   header=None,
                   names=['MovieID', 'Title', 'Genres'],
                   sep='::',engine='python')
movies.head()
```

```python
# 从评分表中读取评分数据
ratings=pd.read_table('DS_day03/code/data/ratings.dat',
                   header=None,
                   names=['UserID', 'MovieID', 'Rating', 'Timestamp'],
                   sep='::',engine='python')
ratings.head()
```
