```python
import pandas as pd
import numpy as np
# 读取数据
data=pd.read_csv('data.csv')
data
```

```python
data.select_dtypes('object').describe().T
```

```python
# 任务一:每一部电影的平均评分
data_title=data.groupby('Title')['Rating'].mean()
data_title
```

```python
# 任务二:同一部电影,不同性别的平均评分
data_gender=data.groupby(['Title','Gender']).mean()
data_gender
```

```python
# 完善任务二:
data_gender_pivot=pd.pivot_table(data=data,index='Title',columns='Gender',values='Rating')
data_gender_pivot
```

```python
# 得到男女分歧最大的前五部电影  np.fabs()
data_gender_pivot['diff']=np.fabs(data_gender_pivot['F']-data_gender_pivot['M'])
data_gender_pivot_sort=data_gender_pivot.sort_values(by='diff',ascending=False)
data_gender_pivot_sort

```

```python
# 任务三:评分最高的十部电影
# 前提: 评论过的次数需要大于100,在参与比较
movies_stats=data.groupby('Title').agg({'Rating':['size','mean']})
# movies_stats
# 定义条件
atleast_100=movies_stats['Rating']['size']>=100
movies_stats_best=movies_stats[atleast_100].sort_values(by=('Rating','mean'),ascending=False)[:10]
movies_stats_best
```
